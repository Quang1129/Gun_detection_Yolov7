### We need to reorganize the train directory to divide it into train and test, in each of which has its own image and label directory.

```python
%cd /content/drive/MyDrive/Yolov7/train_data
!mkdir train
!mkdir train/images
!mkdir train/labels
!mv *.jpg train/images
!mv *.txt train/labels
```

```python
# Declare a yaml file for YOLOv7 to know:
# - Path to the train (if available use the path to train)
# - Number of classes through the variable nc (number of classes)
# - Names of classes
%cd /content/drive/MyDrive/Yolov7/yolov7
!rm data/mydataset.yaml # nếu có
!echo 'train: ../train_data/train' >> data/mydataset.yaml
!echo 'val: ../train_data/train' >> data/mydataset.yaml
!echo 'nc: 2' >> data/mydataset.yaml
!echo "names: ['gun','gun']" >> data/mydataset.yaml
```

```python

%cd /content/drive/MyDrive/Yolov7/yolov7
!python train.py --batch 8 --cfg cfg/training/yolov7.yaml --epochs 8 --data data/mydataset.yaml --weights 'pretrain/yolov7.pt'
```
